在这里我自定义了一个Time_column()函数，其目的是把sleep_wake.csv里的Column“sleep”和Column "Wake"里的数字转化为具体的时刻——“Time”，最后把“Date”加到一起，形成一个DateTime格式。
在这个函数里，column_name 是我么要添加的一列的名字，origin_column就是我们要数据来源的那一列，hours对“Sleep”是22，对“Wake”是6

```python
def Time_column(column_name,origin_column,hours): 
    minute = []
    hour = []
    ##把时间转换一下，-的代表从22：00往回退多少分钟，+的代表增加多少分钟，比如-10代表21：50，15代表22:15
    for i in sleep_data[origin_column]:
        if i > 0:
            minute.append(i - 60*(i//60))
            hour.append(hours + i//60)
        else:
            minute.append(abs(60 + (i-60*(1+i//60))))
            hour.append(hours + i//60)
    list(zip(minute,hour))
    #把日期变成统一的格式，才好切片
    date = [str(i) for i in pd.to_datetime(sleep_data.iloc[:,0])]
    date
    #合并起来，再用pd.datetime()把日期到小时分钟精确起来
    sleep_data[column_name] = [pd.datetime(year=int(i[0][0:4]),month=int(i[0][5:7]),day=int(i[0][8:10]),hour=i[1],minute=i[2]) for i in zip(date,hour,minute)]
```

```python
#给表格增加两列“Wake_time”“Sleep_Time”

Time_column("Wake_Time",'Wake',6)

Time_column("Sleep_Time","Sleep",22)
```

```python
#把表格的index改为datetime格式的“Date"
sleep_data.set_index("Date",inplace=True,drop=True)

sleep_data.index = pd.to_datetime(sleep_data.index)
```

### 创建新的表格

```python
#创建一个以60s为间隔的时间序列
sleep_index = pd.date_range(start=min(sleep_data.index),end=max(sleep_data.index),freq="60S")
#但是我们只要每天21点到23的数据
sleep_index = [time for time in sleep_index if (time.hour >= 21)& (time.hour <= 23)]
sleep_index
```

```python
#创造一个新的dataframe
sleep_frame = pd.DataFrame(0,columns=['indicator'],index=sleep_index)

time_same_date = [i for i in sleep_frame.index if sleep_data['Sleep_Time'][0].date() == i.date()]

time_same_date
```

```python
for Sleep_Time in sleep_data['Sleep_Time']:
    #把日期相同的挑出来
    time_in_day = [i for i in sleep_frame.index if Sleep_Time.date() == i.date()]
    #在日期相同的情况下，把时刻晚于当天sleep_data里的时刻晚的
    time_asleep = [i for i in time_in_day if i.time() >= Sleep_Time.time() ]
    #此处要用loc()，因为loc是按index搜索，iloc()是按序号搜索
    sleep_frame.loc[time_asleep,'indicator'] = 1
sleep_frame
```

下面对清醒的数据如法炮制

```python
#创建一个以60s为间隔的时间序列
awake_index = pd.date_range(start=min(sleep_data.index),end=max(sleep_data.index),freq="60S")
#但是我们只要每天5点到7的数据
awake_index = [time for time in awake_index if (time.hour >= 5)& (time.hour <= 7)]
#创造一个新的dataframe
awake_frame = pd.DataFrame(1,columns=['indicator'],index=awake_index)
```

```python
for Awake_Time in sleep_data['Wake_Time']:
    #把日期相同的挑出来
    time_in_day = [i for i in awake_frame.index if Awake_Time.date() == i.date()]
    #在日期相同的情况下，把时刻晚于当天sleep_data里的时刻晚的
    time_awake = [i for i in time_in_day if i.time() >= Awake_Time.time() ]
    #此处要用loc()，因为loc是按index搜索，iloc()是按序号搜索
    awake_frame.loc[time_awake,'indicator'] = 0
awake_frame
```

接着我们对每个时间点计算它和预期时刻（入睡：22点，起床：6点）的距离

```python
sleep_frame['Time_offset'] = [ int(i.hour)*60 + int(i.minute) - 22*60  for i in sleep_frame.index]
awake_frame['Time_offset'] = [ int(i.hour)*60 + int(i.minute) - 6*60  for i in awake_frame.index]
```
